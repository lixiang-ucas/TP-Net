```python
K = 6
adj = -1*np.ones((len(Edges), 6890,K))
for s in range(len(Edges)):
    edges = Edges[s]
    for i in range(len(edges)):
        a = edges[i,0]
        b = edges[i,1]
        cur = int(np.sum(adj[s,a]!=-1))
#         print(a,b,cur)
        if cur<K:
            adj[s,a,cur] = b
adj += 1  #!!!!!start form 1
adj = adj.astype('int32')   
```
